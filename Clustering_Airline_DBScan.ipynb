```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering, KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.metrics import silhouette_score as sil, calinski_harabasz_score as chs, silhouette_samples
import matplotlib.cm as cm
import scipy.cluster.hierarchy as sch
import warnings
warnings.filterwarnings('ignore')
```

```python
airline = pd.read_excel("EastWestAirlines.xlsx",sheet_name="data")
airline.head()
```

```python
airline.describe()
```

```python
airline.isnull().sum()
```

```python
airline = airline.drop(['ID#'], axis=1)
```

```python
scaler = StandardScaler()
xs = scaler.fit_transform(airline)
xs
```

```python
x_norm = normalize(xs)
x_norm = pd.DataFrame(x_norm ,columns=airline.columns )
x_norm
```

```python
db = DBSCAN(eps=0.3, min_samples=10).fit(x_norm)
labels = db.labels_
```

```python
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print("Number of clusters:" ,n_clusters_)
print("Number of noise points:" , n_noise_)
```

```python
unique_labels = set(labels)
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True

colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = labels == k

    xy = xs[class_member_mask & core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=14,
    )

    xy = xs[class_member_mask & ~core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=6,
    )

plt.title(f"Estimated number of clusters: {n_clusters_}")
plt.show()
```
